```python
import pandas as pd

source_df = pd.read_csv("data\HP_case_data.xlsx - Heat source.csv")
sink_df = pd.read_csv("data\HP_case_data.xlsx - Heat sink.csv")

# Clean up column names (removing trailing spaces or brackets if any)
source_df.columns = source_df.columns.str.strip()
sink_df.columns = sink_df.columns.str.strip()

# Convert time columns to actual Datetime format so they match perfectly
source_df['start measurement'] = pd.to_datetime(source_df['start measurement'])
sink_df['start measurement'] = pd.to_datetime(sink_df['start measurement'])

# Merge the dataframes on the start measurement timestamp
merged_df = pd.merge(source_df, sink_df, on='start measurement', suffixes=('_source', '_sink'))

# Select and rename only the columns we actually need for our TESPy model
cleaned_df = merged_df[[
    'start measurement',
    'T_in[degC]_source',  # Heat source temperature coming in
    'flow[kg/s]',         # Heat source mass flow rate
    'T_in[degC]_sink',    # Heat sink return temperature
    'T_out[degC]',        # Heat sink delivery temperature (90°C)
    'Energy[kWh]'         # The thermal load we need to deliver
]].copy()

# Rename columns to make them easier to code with later
cleaned_df.columns = ['timestamp', 'T_source_in', 'flow_source', 'T_sink_in', 'T_sink_out', 'heat_load_kW']

# Display the first few rows to confirm it looks correct
print(cleaned_df.head())
```
