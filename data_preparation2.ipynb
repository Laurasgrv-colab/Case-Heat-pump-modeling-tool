```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore", category=UserWarning)

source_df = pd.read_excel("data/HP_case_data.xlsx", sheet_name="Heat source")
sink_df = pd.read_excel("data/HP_case_data.xlsx", sheet_name="Heat sink")

# Clean up column names
source_df.columns = source_df.columns.str.strip()
sink_df.columns = sink_df.columns.str.strip()

# Convert time columns to actual Datetime format 
source_df['start measurement'] = pd.to_datetime(source_df['start measurement'])
sink_df['start measurement'] = pd.to_datetime(sink_df['start measurement'])

# Merge the dataframes on the start measurement timestamp
merged_df = pd.merge(source_df, sink_df, on='start measurement', suffixes=('_source', '_sink'))

# Select and rename the columns we need for our TESPy model
cleaned_df = merged_df[[
    'start measurement',
    'T_in[degC_source',  # Heat source temperature coming in
    'flow[kg/s]',         # Heat source mass flow rate
    'T_in[degC_sink',    # Heat sink return temperature
    'T_out[degC]_sink',        # Heat sink delivery temperature (90°C)
    'T_out[degC]_source',
    'Energy[kWh]'         # The thermal load we need to deliver
]].copy()

# Rename columns for comfort
cleaned_df.columns = ['timestamp', 'T_source_in', 'flow_source', 'T_sink_in', 'T_sink_out', 'T_source_out6', 'heat_load_kW']

# Display the first few rows to confirm it looks correct
print(cleaned_df.head())
```
