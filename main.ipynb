```python
import pandas as pd

from src.config import CONFIG
from src.heatpump import HeatPumpModel
#from src.model3 import HeatPumpModel
from results.plotting import HeatPumpPlotter

# Read dataset
df = pd.read_excel(
    "data/HP_case_data.xlsx",
    sheet_name="Heat source"
)

# Build model
hp = HeatPumpModel(CONFIG)

hp.build()

hp.design()

results = []

for _, row in df.iterrows():

    hp.update_boundary_conditions(row)

    hp.offdesign()

    results.append(hp.get_results())

results = pd.DataFrame(results)

# Plot
plotter = HeatPumpPlotter(results)

plotter.plot_all()
```
